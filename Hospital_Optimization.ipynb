**5.1  Remove rows where BOTH dates are missing**

```python
admissions_clean = admissions_clean.dropna(subset=['d_o_a'], how='any')
```

**STEP 5.2  If discharge date is missing  fill using admission + duration**

```python
missing_discharge = admissions_clean['d_o_d'].isna()

admissions_clean.loc[missing_discharge, 'd_o_d'] = (
    admissions_clean.loc[missing_discharge, 'd_o_a'] +
    pd.to_timedelta(admissions_clean.loc[missing_discharge, 'duration_of_stay'], unit='D')
)
```

STEP 5.3  Recalculate duration_of_stay (ensure correct)

```python
admissions_clean['computed_stay'] = (
    admissions_clean['d_o_d'] - admissions_clean['d_o_a']
).dt.days
```

```python
admissions.head()
```

6.1  Standardize gender

```python
admissions['gender'] = admissions['gender'].astype(str).str.strip().str.upper()

admissions['gender'] = admissions['gender'].replace({
    'M': 'Male',
    'F': 'Female'
})
```

6.2  Standardize rural/urban

```python
admissions['rural'] = admissions['rural'].astype(str).str.strip().str.upper()

admissions['rural'] = admissions['rural'].replace({
    'R': 'Rural',
    'U': 'Urban'
})
```

6.3  Standardize admission type

```python
admissions['type_of_admission_emergency_opd'] = (
    admissions['type_of_admission_emergency_opd']
    .astype(str)
    .str.strip()
    .str.upper()
)

admissions['type_of_admission_emergency_opd'] = admissions['type_of_admission_emergency_opd'].replace({
    'E': 'Emergency',
    'O': 'OPD'
})
```

6.4  Standardize outcome

```python
admissions['outcome'].unique()
```

```python
admissions['outcome'] = admissions['outcome'].replace({
    'DISCHARGE': 'Discharged',
    'EXPIRY': 'Death',
    'DAMA': 'Left_Against_Medical_Advice'
})
```

```python
admissions[['gender', 'rural', 'type_of_admission_emergency_opd', 'outcome']].head()
```

```python
admissions['d_o_a'] = pd.to_datetime(admissions['d_o_a'], format='%m-%d-%Y', errors='coerce')
admissions['d_o_d'] = pd.to_datetime(admissions['d_o_d'], format='%m-%d-%Y', errors='coerce')
```

```python
admissions[['d_o_a', 'd_o_d']].head()
```

```python
# Day of week
admissions['day_of_week'] = admissions['d_o_a'].dt.dayofweek

# Month number
admissions['month'] = admissions['d_o_a'].dt.month

# Year
admissions['year'] = admissions['d_o_a'].dt.year

# Weekend flag
admissions['is_weekend'] = admissions['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# Stay length
admissions['stay_length'] = (admissions['d_o_d'] - admissions['d_o_a']).dt.days

# Day of year
admissions['day_of_year'] = admissions['d_o_a'].dt.dayofyear
```
